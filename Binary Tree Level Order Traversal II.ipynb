### [Binary Tree Level Order Traversal II](https://leetcode.com/problems/binary-tree-level-order-traversal-ii/description/)

Given a binary tree, return the bottom-up level order traversal of its nodes' values. (ie, from left to right, level by level from leaf to root).

For example:
Given binary tree [3,9,20,null,null,15,7],
```
    3
   / \
  9  20
    /  \
   15   7
```
return its bottom-up level order traversal as:
```
[
  [15,7],
  [9,20],
  [3]
]
```

````python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

from collections import deque

class Solution:
    def levelOrderBottom(self, root):
        """
        :type root: TreeNode
        :rtype: List[List[int]]
        """
        return self.levelOrderBottomBFSv2(root)
    
    def levelOrderBottomBFSv2(self, root):
        """
        :type root: TreeNode
        :rtype: List[List[int]]
        """
        
        if not root:
            return []
        
        queue = deque()
        bottomUp = deque()
        
        queue.append(root)
        
        while queue:
            count = len(queue)
            nodesAtCurrentLevel = []
            for _ in range(count):
                node = queue.popleft()
                nodesAtCurrentLevel.append(node.val)
                if node.left:
                    queue.append(node.left)
                if node.right:
                    queue.append(node.right)
            
            bottomUp.appendleft(nodesAtCurrentLevel)
        
        return list(bottomUp)
            
            
        
    def levelOrderBottomBFSv1(self, root):
        """
        :type root: TreeNode
        :rtype: List[List[int]]
        """
        # can use a queue, because we go bottom up.
        # first in first out.
        
        # edge cases
        if not root:
            return []
        
        queue = deque()
        valuesInBottomUpOrder = deque()
        
        queue.append((root, 1))
        
        while queue:
            node, level = queue.popleft()
            
            # process the node.. add it to the bottom up list
            if len(valuesInBottomUpOrder) < level:
                valuesInBottomUpOrder.appendleft([node.val])
            else:
                valuesInBottomUpOrder[0].append(node.val)
            
            if node.left:
                queue.append((node.left, level+1))
            
            if node.right:
                queue.append((node.right, level+1))
        
        # Conversion from deque to list is needed to pass the grader.
        # In actual implementation, caller can work with the deque itself.
        return list(valuesInBottomUpOrder)
        
````
