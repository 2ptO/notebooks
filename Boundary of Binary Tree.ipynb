### [Boundary of Binary Tree](https://leetcode.com/problems/boundary-of-binary-tree/description/)

Given a binary tree, return the values of its boundary in anti-clockwise direction starting from root. Boundary includes left boundary, leaves, and right boundary in order without duplicate nodes.

Left boundary is defined as the path from root to the left-most node. Right boundary is defined as the path from root to the right-most node. If the root doesn't have left subtree or right subtree, then the root itself is left boundary or right boundary. Note this definition only applies to the input binary tree, and not applies to any subtrees.

The left-most node is defined as a leaf node you could reach when you always firstly travel to the left subtree if exists. If not, travel to the right subtree. Repeat until you reach a leaf node.

The right-most node is also defined by the same way with left and right exchanged.

Example 1
```
Input:
  1
   \
    2
   / \
  3   4

Ouput:
[1, 3, 4, 2]

Explanation:
The root doesn't have left subtree, so the root itself is left boundary.
The leaves are node 3 and 4.
The right boundary are node 1,2,4. Note the anti-clockwise direction means you should output reversed right boundary.
So order them in anti-clockwise without duplicates and we have [1,3,4,2].
```
Example 2
```
Input:
    ____1_____
   /          \
  2            3
 / \          / 
4   5        6   
   / \      / \
  7   8    9  10  
       
Ouput:
[1,2,4,7,8,9,10,6,3]

Explanation:
The left boundary are node 1,2,4. (4 is the left-most node according to definition)
The leaves are node 4,7,8,9,10.
The right boundary are node 1,3,6,10. (10 is the right-most node).
So order them in anti-clockwise without duplicate nodes we have [1,2,4,7,8,9,10,6,3].
```

````python
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None

from collections import deque

class Solution:
    def boundaryOfBinaryTree(self, root):
        """
        :type root: TreeNode
        :rtype: List[int]
        """
        # Logic is almost same as the brute force.
        # with slight optimizations.
        
        if not root:
            return []
        
        def isLeaf(node):
            return (node and not node.left and not node.right)
        
        def collectLeftBoundary(node, leftBoundary):
            """
            Collects the values of the nodes on the left boundary.
            Leaf nodes are excluded.
            """
            # Get the path to the left most nodes
            while node:
                if not isLeaf(node):
                    leftBoundary.append(node.val)
                if node.left:
                    node = node.left
                else:
                    node = node.right
                    
        def collectRightBoundary(node, rightBoundary):        
            # Get the path to the right most nodes
            while node:
                if not isLeaf(node):
                    rightBoundary.appendleft(node.val)
                if node.right:
                    node = node.right
                else:
                    node = node.left

        def collectLeaves(node, leaves):
            if node:
                collectLeaves(node.left, leaves)
                if isLeaf(node):
                    leaves.append(node.val)
                collectLeaves(node.right, leaves)
        
        if isLeaf(root):
            return [root.val]
                    
        leftBoundary = [root.val]
        rightBoundary = deque() # Using deque due to anti-clockwise order
        leaves = []
        
        # start from left subtree
        collectLeftBoundary(root.left, leftBoundary)
        
        # Go right
        collectRightBoundary(root.right, rightBoundary)
        
        # Collect the leaves. It is not the leaves at the last level. All leaves
        # should be considered
        # Doing a DFS to collect the leaves
        collectLeaves(root, leaves)
        
        print(leftBoundary, leaves, rightBoundary)
        return leftBoundary + leaves + list(rightBoundary)
        
    def boundaryOfBinaryTreeBruteForce(self, root):
        """
        :type root: TreeNode
        :rtype: List[int]
        """
         # get the path to the left most node
        # get the path to the right most node without the root
        
        if not root:
            return []
        
        def isLeaf(node):
            return (node and not node.left and not node.right)
        
        if isLeaf(root):
            return [root.val]
        
        # start from left subtree
        node = root.left
        leftBoundary = [root.val]
        leftLeaves = 0
        
        # Get the path to the left most nodes
        while node:
            leftBoundary.append(node.val)
            if isLeaf(node):
                leftLeaves += 1
            if node.left:
                node = node.left
            else:
                node = node.right
        
        # Get the path to the right most nodes
        rightBoundary = deque()
        node = root.right
        rightLeaves = 0
        while node:
            rightBoundary.appendleft(node.val)
            if isLeaf(node):
                rightLeaves += 1
            if node.right:
                node = node.right
            else:
                node = node.left
        
        # Collect the leaves. It is not the leaves at the last level. All leaves
        # should be considered
        # Doing a DFS to collect the leaves
        leaves = []
        def dfs(node, leaves):
            if node:
                dfs(node.left, leaves)
                # Is node a leaf node?
                if not node.left and not node.right:
                    leaves.append(node.val)
                dfs(node.right, leaves)
        
        dfs(root, leaves)
        
        # Exclude the values of leaves at the left and right most ends as they
        # would have already counted in the left or right tree.
        print(leftBoundary, leaves, rightBoundary)
        
        return leftBoundary + leaves[leftLeaves:len(leaves) - rightLeaves] + list(rightBoundary)
        # else:
        #     return leftBoundary + list(rightBoundary)
````
