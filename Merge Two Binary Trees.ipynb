### [Merge Two Binary Trees](https://leetcode.com/problems/merge-two-binary-trees/description/)



```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def mergeTrees(self, t1, t2):
        """
        :type t1: TreeNode
        :type t2: TreeNode
        :rtype: TreeNode
        """
        # merge two trees
        # traverse both the trees, starting from root
        # 
        
        # t1 and t2 must have a root. 
        if not t1 or not t2:
            return t1 or t2
        
        # Modify the tree in place. Use either t1 or t2 as the third tree.
        # This solution is relatively easier
        t1.val += t2.val
        
        t1.left = self.mergeTrees(t1.left, t2.left)
        t1.right = self.mergeTrees(t1.right, t2.right)
        
        return t1
    
    def mergeTreesIntoNew(self, t1, t2):
        """
        :type t1: TreeNode
        :type t2: TreeNode
        :rtype: TreeNode
        """
        # Pretty much same time complexity as the previous, but allocates
        # a new tree for the sake of correctness
        
        def mergeTwo(t1, t2):
            if t1 and t2:
                return TreeNode(t1.val + t2.val)
            elif (t1 or t2):
                if t1:
                    return TreeNode(t1.val)
                else:
                    return TreeNode(t2.val)
        
        def mergeV2(t1, t2, t3):
            if t1 and t2:
                t3.left = mergeTwo(t1.left, t2.left)
                t3.right = mergeTwo(t1.right, t2.right)
                
                merge(t1.left, t2.left, t3.left)
                merge(t1.right, t2.right, t3.right)
                
            elif t1 or t2:
                if t1:
                    t3.left = mergeTwo(t1.left, None)
                    t3.right = mergeTwo(t1.right, None)
                    merge(t1.left, None, t3.left)
                    merge(t1.right, None, t3.right)
                else:
                    t3.left = mergeTwo(t2.left, None)
                    t3.right = mergeTwo(t2.right, None)
                    merge(t2.left, None, t3.left)
                    merge(t2.right, None, t3.right)
        
        # t1 and t2 must have a root. 
        if not t1 or not t2:
            return t1 or t2
        
        t3 = TreeNode(t1.val + t2.val)
        merge(t1, t2, t3)
        
        return t3
```
