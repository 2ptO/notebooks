### [Min Stack](https://leetcode.com/problems/min-stack/)

Design a stack that supports push, pop, top, and retrieving the minimum element in constant time.

push(x) -- Push element x onto stack.
pop() -- Removes the element on top of the stack.
top() -- Get the top element.
getMin() -- Retrieve the minimum element in the stack.

```
Example:
MinStack minStack = new MinStack();
minStack.push(-2);
minStack.push(0);
minStack.push(-3);
minStack.getMin();   --> Returns -3.
minStack.pop();
minStack.top();      --> Returns 0.
minStack.getMin();   --> Returns -2.
```

````python
class MinStack(object):

    def __init__(self):
        """
        initialize your data structure here.
        """
        # Notes
        
        # push() - O(1)
        # getMin() - O(1)
        # pop() - trade off here???
        #    update stack with next smallest element.
        #    so track the min elements only.
        #    how to track them? - auxilary stack?
        # main_stack: 1 4 5 | aux_stack: 1
        # 1 4 5 0   | aux_stack: 1 0
        # 1 4 5 0 -2 | aux_stack 1 0 -2
        # 1 4 5 0 -2 4 | aux_stack: 1 0 -2
        # - if the popped element is the lowest element,
        # then pop the item from aux stack as well.
        # space: O(n) in worst case
        # edge cases: empty stack, empty min stack
        # assume: no stack limit, so no overflow
        
        self.stack = [] # main stack
        self.minStack = [] # keep track of min values
        

    def push(self, x):
        """
        :type x: int
        :rtype: void
        """
        # add to main stack
        self.stack.append(x)
        
        # add to min stack if it is less than or equal to the
        # top of min stack or min stack is empty
        if not self.minStack or x <= self.minStack[-1]:
            self.minStack.append(x)
        

    def pop(self):
        """
        :rtype: void
        """
        if not self.stack:
            return
        
        # if main stack is not empty, then minstack should also be not empty
        assert len(self.minStack) > 0, "Min stack should not be empty"
        
        popped = self.stack.pop()
        
        if popped == self.minStack[-1]:
            # popped value is the lowest.
            # update min stack as well.
            self.minStack.pop()
        

    def top(self):
        """
        :rtype: int
        """
        if not self.stack:
            return None
        
        return self.stack[-1]

    def getMin(self):
        """
        :rtype: int
        """
        if not self.minStack:
            return None
        
        return self.minStack[-1]
        



# Your MinStack object will be instantiated and called as such:
# obj = MinStack()
# obj.push(x)
# obj.pop()
# param_3 = obj.top()
# param_4 = obj.getMin()
````
