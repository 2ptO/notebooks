### [Insert Delete GetRandom in O(1)](https://leetcode.com/problems/insert-delete-getrandom-o1/)

Design a data structure that supports all following operations in average O(1) time.

- insert(val): Inserts an item val to the set if not already present. 
- remove(val): Removes an item val from the set if present.
- getRandom: Returns a random element from current set of elements. Each element must have the same probability of being returned.

Example:
```
// Init an empty set.
RandomizedSet randomSet = new RandomizedSet();

// Inserts 1 to the set. Returns true as 1 was inserted successfully.
randomSet.insert(1);

// Returns false as 2 does not exist in the set.
randomSet.remove(2);

// Inserts 2 to the set, returns true. Set now contains [1,2].
randomSet.insert(2);

// getRandom should return either 1 or 2 randomly.
randomSet.getRandom();

// Removes 1 from the set, returns true. Set now contains [2].
randomSet.remove(1);

// 2 was already in the set, so return false.
randomSet.insert(2);

// Since 2 is the only number in the set, getRandom always return 2.
randomSet.getRandom();
```

````python
import random

class RandomizedSet(object):

    def __init__(self):
        """
        Initialize your data structure here.
        """
        # O(1) time for
        #   insert
        #   remove
        #   getRandom - each element returned should have the same probability
        
        # we are given only one val.. not key..val pair.
        #
        # what data structure can support this?
        #   hash, set, map..???
        #
        # hash/map - insertion and deletion operates in O(1) at amortized cost
        #   list.. to store the values.. addition to list is O(1)
        #   hashtable - that maps the val and its position in the list
        # insert
        #   if the val already in the hashmap
        #       # if we are to allow duplicates, then this changes the scene
        #       return False
        #   otherwise, add to our list and to our map
        #   not bounded by size. so keep collision outside for now.
        # delete
        #   if the valu not in our map
        #       return False
        #   get the index in the list
        #   if we know the index, then deletion is O(1) time
        #   deletion will alter the index of the remaining items in the list.
        #   how can we handle this?
        #       move the last item in the list to the deleted position
        #       and update index of the last item to deleted position
        #   
        # get random
        #   get a random number within the range of container size.
        
        self.container = []
        self.containerMap = {}
        

    def insert(self, val):
        """
        Inserts a value to the set. Returns true if the set did not already contain the specified element.
        :type val: int
        :rtype: bool
        """
        if val in self.containerMap:
            return False
        
        self.container.append(val)
        self.containerMap[val] = len(self.container) - 1
        
        return True
        
    def remove(self, val):
        """
        Removes a value from the set. Returns true if the set contained the specified element.
        :type val: int
        :rtype: bool
        """
        
        if val not in self.containerMap:
            return False
        
        # make sure list is not empty
        # if val is in containerMap then list should also be not empty
        valIndex = self.containerMap[val]
        
        # move the last element to the deleted position
        self.container[valIndex] = self.container[-1]
        self.containerMap[self.container[-1]] = valIndex
        
        # drop the last element.
        self.container.pop()
        
        # remove the deleted val from our map as well
        del self.containerMap[val]
        
        return True

    def getRandom(self):
        """
        Get a random element from the set.
        :rtype: int
        """
        if self.container:
            # return random.choice(self.container)
            return self.container[random.randrange(len(self.container))]
        else:
            raise IndexError


# Your RandomizedSet object will be instantiated and called as such:
# obj = RandomizedSet()
# param_1 = obj.insert(val)
# param_2 = obj.remove(val)
# param_3 = obj.getRandom()
````
