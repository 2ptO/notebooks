### [Count Complete Tree Nodes](https://leetcode.com/problems/count-complete-tree-nodes/description/)

Given a complete binary tree, count the number of nodes.

Note:

Definition of a complete binary tree from Wikipedia:
In a complete binary tree every level, except possibly the last, is completely filled, and all nodes in the last level are as far left as possible. It can have between 1 and 2<sup>h</sup> nodes inclusive at the last level h.

**Example:**

```
Input: 
    1
   / \
  2   3
 / \  /
4  5 6

```
Output: 6

````python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def countNodes(self, root):
        """
        :type root: TreeNode
        :rtype: int
        """
         # counting the number of nodes
        # 
        # complete tree: all nodes in h-1 level would be filled.
        # in the level h, nodes in the last level are filled as far left as possible
        #
        # is the tree guaranteed to be complete?
        # should I check for that?
        # 
        # we can do a BFS.. count every node... O(n)
        # 
        # There must be a way to do this less than O(n)..using the completeness property
        #
        # find the height. 2^h-1 - 1 nodes are guaranteed. if we go down left most.. that will
        # give the height.. O(h) steps.
        # left most in the last level.. right most in the last level. diff + 1 + numNodesInPrevLevel
        # should give the answer.
        # left most can be found easily. challenge is how to find the right most in the last level?
        # that could be in the left subtree or right subtree.
        # split the problem here.. count the number of nodes in the left subtree and right separately
        # if left and right subtree are in same height, then last node must be in the right subtree
        # else: last node must be in the left subtree
        
        if not root:
            return 0
        
        count = 0
        while root:
            rootHeight = self.getHeight(root) # this is not really needed. instead we can simply reduce the height by 1
            rightHeight = self.getHeight(root.right)

            if rightHeight == rootHeight - 1: # left and right subtree are in the same level
                # left tree must also be complete. left Height = h - 1
                # so number of nodes = 2^h - 1 + 1 (for root) = 2^h
                # look for the last node in the last level in the right subtree
                count += (2 ** rootHeight) # height using 0 indexing.. num nodes in subtree = 2^(h) - 1
                root = root.right
            else:
                # right tree is complete here.. so count the nodes here
                count += (2 ** (rootHeight - 1)) 
                # look for the last node in the last level in the left subtree
                root = root.left
        
        return count

        
    def getHeight(self, node):
        height = -1
        while node:
            node = node.left
            height += 1
        return height
        
````
