### [Number to English Words](https://leetcode.com/problems/integer-to-english-words/description/)

Convert a non-negative integer to its english words representation. Given input is guaranteed to be less than 231 - 1.

Example 1:
```
Input: 123
Output: "One Hundred Twenty Three"
```
Example 2:
```
Input: 12345
Output: "Twelve Thousand Three Hundred Forty Five"
```
Example 3:
```
Input: 1234567
Output: "One Million Two Hundred Thirty Four Thousand Five Hundred Sixty Seven"
```
Example 4:
```
Input: 1234567891
Output: "One Billion Two Hundred Thirty Four Million Five Hundred Sixty Seven Thousand Eight Hundred Ninety One"
```

````python
from collections import deque
class Solution:
    def numberToWords(self, num):
        """
        :type num: int
        :rtype: str
        """
        # cleaned up the first attempt with better formatting
        
        # Edge cases
        if num < 0 or num > ((1 << 31) - 1):
            # invalid range
            return ""
        
        # other edge cases
        if num == 0:
            return "Zero"
        
        numStr = str(num)
        # set numStr to multiples of 3 in length for easy processing
        if len(numStr) % 3 != 0:
            numStr = ("0" * (3 - len(numStr) % 3)) + numStr
        
        ones = [
            "", 
            "One",
            "Two",
            "Three", 
            "Four",
            "Five", 
            "Six", 
            "Seven", 
            "Eight", 
            "Nine"
        ]
        
        teens = [
            "Ten", 
            "Eleven", 
            "Twelve", 
            "Thirteen", 
            "Fourteen", 
            "Fifteen", 
            "Sixteen", 
            "Seventeen", 
            "Eighteen", 
            "Nineteen"
        ]
        
        tenths = [
            "", 
            "Ten", 
            "Twenty",
            "Thirty", 
            "Forty", 
            "Fifty", 
            "Sixty", 
            "Seventy",
            "Eighty", 
            "Ninety"
        ]
        
        depthsToWord = [
            "", 
            "Thousand",
            "Million", 
            "Billion"
        ]
        
        depth = 0
        words = ""
        
        for i in range(len(numStr)-1, -1, -3): # Stepping three digits at a time
            curNum = numStr[i-2:i+1]
            curWord = deque()
            
            # add second digit and first digit together if they
            # together form the special teen value
            if curNum[1] == "1": # pick from teens
                curWord.appendleft(teens[int(curNum[2])])
            else:
                # add first digit
                if ones[int(curNum[2])]:
                    curWord.appendleft(ones[int(curNum[2])])
                    
                # add second digit
                if int(curNum[1]) > 0:
                    curWord.appendleft(tenths[int(curNum[1])])
                    
            # add third digit
            if int(curNum[0]) > 0:
                curWord.appendleft(ones[int(curNum[0])] + " Hundred")
            
            # if a valid curword formed in this iteration, then add it to 
            # total word.
            if curWord:
                words = " ".join(curWord) + " " + depthsToWord[depth] + " " + words
            
            # Go to next depth
            depth += 1
        
        # remove leading and trailing white spaces to make the OJ happy
        return words.lstrip().rstrip()
           

    def numberToWordsBruteForce(self, num):
        """
        :type num: int
        :rtype: str
        """
        
#         break into chunks of 3 <- depth level determines 0 - none, 1 - thousand, 2 - million, 3 - billion
#         .. within that break into ones, tens,  hundred
#         .. s = '567'
#         .. s[2] = mapped from .. {one, two, three...}
#         .. s[1] = mapped from {ten, twenty, thirty, forty, fifty, sixty, ...}
#         ..s[0] = mapped from {one, two, three..} + hundred
#         .. form the string representation for the 3 digit
#         ..add the string representation of the depth level 
#         fullform = current rep + full form

#         given input is number.. not a list or string.
#         trying to solve it with string first.. number will require some
#         math.. for e.g to get the last three digits .. we have to take mod
#         of 1000
#         
        # what about zeros?
        # 60
        # 600
        # 606 if 0 don't add any text.. simply ""
        # 60 -> sixty for 6, "" for 0
        # 61
        # 001 -> one
        # 
        # three maps..rather list is fine too.
        #
        # ones, tens .. can add hundred to ones
        # so two is enough
        
        # edge cases
        if num < 0 or num > ((1 << 31) - 1):
            # invalid range
            return ""
        
        # other edge cases
        # 0
        # 0..num.. leading zeros don't count.
        # trailing zeroes do.
        if num == 0:
            return "Zero"
        
        numStr = str(num)
        # set numStr to multiples of 3 in length for easy processing
        if len(numStr) % 3 != 0:
            numStr = ("0" * (3 - len(numStr) % 3)) + numStr
        
        print(numStr)
        
        ones = ["", "One", "Two", "Three", "Four", "Five", "Six", "Seven", "Eight", "Nine"]
        # break. seem to be going wrong with tens
        # 11 -> Ten One - thats wrong.. should say Eleven
        # 12 -> Twelve, not Two One
        # 412 -> four hundred twelve, not Four Hundred Ten two
        # teens.. 10.. 19 are special cases
        # if second digit is 1.. then it is a special case
        # if not pull from tens
        teens = ["Ten", "Eleven", "Twelve", "Thirteen", "Fourteen", "Fifteen", "Sixteen", "Seventeen", "Eighteen", "Nineteen"]
        tenths = ["", "Ten", "Twenty", "Thirty", "Forty", "Fifty", "Sixty", "Seventy", "Eighty", "Ninety"]
        
        depthsToWord = ["", "Thousand", "Million", "Billion"]
        depth = 0
        words = ""
        for i in range(len(numStr)-1, -1, -3): # Stepping three digits at a time
            curNum = numStr[i-2:i+1]
            curWord = deque()
            
            # add second digit and first digit together if they
            # together form the special teen value
            if curNum[1] == "1": # pick from teens
                curWord.appendleft(teens[int(curNum[2])])
            else:
                # add first digit
                if ones[int(curNum[2])]:
                    curWord.appendleft(ones[int(curNum[2])])
                    
                # add second digit
                if int(curNum[1]) > 0:
                    curWord.appendleft(tenths[int(curNum[1])])
                    
            # add third digit
            if int(curNum[0]) > 0:
                curWord.appendleft(ones[int(curNum[0])] + " Hundred")
            
            # if a valid curword formed in this iteration, then add it to 
            # total word.
            if curWord:
                words = " ".join(curWord) + " " + depthsToWord[depth] + " " + words
            depth += 1
        
        # remove leading and trailing white spaces to make the OJ happy
        return words.lstrip().rstrip()
        
        
````

````python
testCases = [
    (123, "One Hundred Twenty Three"),
    (0, "Zero"),
    (1, "One"),
    (11, "Eleven"),
    (101, "One Hundred One"),
    (1000, "One Thousand"),
    (1234604, "One Million Two Hundred Thirty Four Thousand Six Hundred Four"),
    (141234604, "One Hundred Forty One Million Two Hundred Thirty Four Thousand Six Hundred Four"
),
    (2141234604, "Two Billion One Hundred Forty One Million Two Hundred Thirty Four Thousand Six Hundred Four"
)
]

s = Solution()

for testInput, expOutput in testCases:
    assert s.numberToWords(testInput) == expOutput
````
