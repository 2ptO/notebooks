### [Reverse Words In a String](https://leetcode.com/problems/reverse-words-in-a-string/)


Given an input string, reverse the string word by word.

Example:  

```
Input: "the sky is blue",
Output: "blue is sky the".
```
Note:

- A word is defined as a sequence of non-space characters.
- Input string may contain leading or trailing spaces. However, your reversed string should not contain leading or trailing spaces.
- You need to reduce multiple spaces between two words to a single space in the reversed string.


````python
class Solution(object):
    def reverseWords(self, s):
        """
        :type s: str
        :rtype: str
        """
        
        # reverse the string word by word
        # 
        # definition of word:
        #   sequence of non-space characters (any char set it may be)
        # input string may contain leading or trailing spaces
        # no such leading or trailing spaces in the final output.
        
        # it is reversal of entire string per se. reversal of words within a sentence.
        #
        # brute force:
        #   reverse the entire sentence.
        #   reverse each word within the sentence
        # 
        # e.g. the sky is blue
        #       step1: eulb si yks eht
        #       step2: blue is sky the
        # we walk the sentence twice. 
        # can we do better?
        # 
        # split the words and join in reverse order
        # walk from the end then? 
        
        # start from the end.. walk until a space is hit
        # if space and previous start == non-space:
        #   newword = s[space+1: start]
        #   reversedString += newWord
        
        # edge cases
        #   string is empty
        #   string of length 1
        
        
        if not s:
            return s
        
        index = len(s) - 1
        wordEnd = -1
        
        # can also use a list and join the individual words at the end
        reversedWords = ""
        
        while index >= 0:
            if s[index].isspace():
                if wordEnd != -1:
                    # hit a split point
                    # calculate the start, grab the new word and add it to our list.
                    wordStart = index+1
                    
                    # want end character to be included as well
                    newWord = s[wordStart:wordEnd+1]
                    
                    reversedWords += newWord + " "

                    # reset the end to search for next word
                    wordEnd = -1
                else:
                    # unwanted space.
                    # skip it.
                    pass
                
            else:
                if wordEnd == -1:
                    wordEnd = index
            
            index -= 1
        
        if wordEnd != -1: # we still have unmatched word. Grab that too.
            reversedWords += s[:wordEnd+1] + " "

        return reversedWords[:-1] if reversedWords else ""
````

````python
# some basic test cases
s = Solution()
testStringMap = {
    "the sky is blue   ": "blue is sky the",
    "string is empty " : "empty is string",
    "    " : "",
    " quick brown fox ": "fox brown quick"
}

for testString in testStringMap:
    expOutput = testStringMap[testString]
    assert(s.reverseWords(testString) == expOutput)
    
````
