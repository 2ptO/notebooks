### [Word Search](https://leetcode.com/problems/word-search/description/)

Given a 2D board and a word, find if the word exists in the grid.

The word can be constructed from letters of sequentially adjacent cell, where "adjacent" cells are those horizontally or vertically neighboring. The same letter cell may not be used more than once.

Example:

```
board =
[
  ['A','B','C','E'],
  ['S','F','C','S'],
  ['A','D','E','E']
]

Given word = "ABCCED", return true.
Given word = "SEE", return true.
Given word = "ABCB", return false.
```

````python
class Solution(object):
    def exist(self, board, word):
        """
        :type board: List[List[str]]
        :type word: str
        :rtype: bool
        """
        # backtracking
        #   cell matches with a char.. next char can be in four possible places
        #   one on the left, right, above and below
        #   if at least one match is found, move in that direction. but hold on.
        #   if moving in that direction, finds a word then its ok..return true.
        #   if it doesn't find a word, then find if another match is there. 
        # if cell doesn't match, then move horizontally
        # if cell match, but next char doesn't match then ? look for another start? look for whole word again
        # 
        # I have to construct the word from the board
        # at each cell, check wehther the constructed word == word, return True
        #
        # check for the formation of the word from every cell // need two for loops
        # check function separate.. that probably have to use recursion
        #
        # what if I come to the same cell again?
        # // need to keep track of cell which are already visited
        
        # edge cases 
        if not board:
            return False
        
        if not word:
            return True
        
        def checkWord(board, word, wordPos, row, column, visitedCells):
            if wordPos == len(word):
                # we found all the characters
                return True
            
            # check for row and column boundaries
            if row < 0 or row >= len(board):
                return False
            
            if column < 0 or column >= len(board[0]):
                return False
            
            # if visited this already, quit.
            if visitedCells[row][column]:
                return False
            
            # if cell doesn't match the current character in word,
            # then also quit
            if word[wordPos] != board[row][column]:
                return False
            
            # ok.. we passed all the checks.
            # now check the neighbors
            # but before that, mark this cell as visited to avoid
            # infinite recursion.b
            
            visitedCells[row][column] = True
            
             
            cellNext = checkWord(board, word, wordPos+1, row, column + 1, visitedCells)
            # Do not check additonal cells unnecessarily.
            cellPrev = cellNext or checkWord(board, word, wordPos+1, row, column - 1, visitedCells)
            cellAbove = cellPrev or checkWord(board, word, wordPos+1, row - 1, column, visitedCells)
            cellBelow = cellAbove or checkWord(board, word, wordPos+1, row + 1, column, visitedCells)
            
            # Turn off visited for next iteration
            visitedCells[row][column] = False
            
            # if we found the remainder of the word from any of the adjacent
            # cells, return True
            return (cellNext or cellPrev or cellAbove or cellBelow)
        
        
        # 2-D board. Get the rows and columns.
        rows = len(board)
        columns = len(board[0])
        visitedCells = [[False for _ in range(columns)] for _ in range(rows)] # Additional space
        
        # Say board has m-rows and n-columns
        for row in range(rows): # O(m)
            for column in range(columns): # O(n)
                if checkWord(board, word, 0, row, column, visitedCells): # O(mn) in worst case
                    return True
                # if word doesn't exist from current cell, move to next cell
        
        return False
    
        # Time complexity: O(m^2 * n^2)
        # Space: O(mn)
                
                
````

````python
from typing import List
class Solution:
    def exist(self, board: List[List[str]], word: str) -> bool:
        # given 2-D board, where each cell has a char,
        # and a word, find if the word exists in the board
        
        # constraints to construct the word:
        #   sequentially adjacent
        #       horizontally or vertically
        #       can go in 4 different combinations at each cell
        
        # start from the cell with the first char of the word
        #   coz, word has to be constructed sequentially
        #
        # can go in 4 directions..mark the visited cells along the way
        #   pick the direction which has the next char.
        #   exist(board, BCCED)
        
        # walk down the board until first char is met
        #   dfs()
        
        def dfs(board, row, col, word, visited, found):
            n_rows = len(board)
            n_cols = len(board[0])
            
            visited[row][col] = True
            
            # can go in the direction of next char in line
            if len(word) <= 1:
                found[0] = True
                return
            
            next_char = word[1]
            neighbors = [(row, col-1), (row-1, col), (row, col+1), (row+1, col)]
            for neighbor in neighbors:
                i, j = neighbor
                
                if ((0 <= i < n_rows) and 
                    (0 <= j < n_cols) and
                    (board[i][j] == next_char) and
                    not visited[i][j]): 
                    dfs(board, i, j, word[1:], visited, found)
            
            visited[row][col] = False
        if not word:
            # assuming empty word can be formed from any board
            return True
        
        n_rows = len(board)
        n_cols = len(board[0])
        
        visited = [[False] * n_cols for _ in range(n_rows)]
        found = [False]
        
        for i in range(n_rows):
            for j in range(n_cols):
                if board[i][j] == word[0]:
                    dfs(board, i, j, word, visited, found)
                    if found[0]:
                        return True
        
        return False
    
#     [
#         ["C","A","A"],
#         ["A","A","A"],
#         ["B","C","D"]
#     ]
# "AAB"
````
