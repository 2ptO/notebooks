### [Binary Tree Pruning](https://leetcode.com/problems/binary-tree-pruning/)

We are given the head node root of a binary tree, where additionally every node's value is either a 0 or a 1.

Return the same tree where every subtree (of the given tree) not containing a 1 has been removed.

(Recall that the subtree of a node X is X, plus every node that is a descendant of X.)

```
Example 1:
Input: [1,null,0,0,1]
Output: [1,null,0,null,1]

Explanation: 
Only the red nodes satisfy the property "every subtree not containing a 1".
The diagram on the right represents the answer.
```



```
Example 2:
Input: [1,0,1,0,0,0,1]
Output: [1,null,1,null,1]
```


```
Example 3:
Input: [1,1,0,1,1,0,1,0]
Output: [1,1,0,1,1,null,1]
```


Note:

The binary tree will have at most 100 nodes.
The value of each node will only be 0 or 1.


````python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def pruneTree(self, root):
        """
        :type root: TreeNode
        :rtype: TreeNode
        """
        
        # subtree - root + left and right
        # prune if there is no '1' in the subtree
        #   root = 0 and (no left or left == 0) and (no right or right == 0)
        # go to use recursion.. so some sort of depth-first.. preferably post-order
        # root.left = root.left if not prune(root.left) else None
        # root.right = root.right if not prune(root.right) else None
        # 
        # returns?? True or False??
        # if left or right exists and either of them is 1 return False else True
        # is root always guaranteed to be 1?
        
        def prune(root):
            if not root:
                return True
            
            leftPruned = prune(root.left)
            rightPruned = prune(root.right)
            
            root.left = None if leftPruned else root.left
            root.right = None if rightPruned else root.right
            
            if (root.val == 0 and leftPruned and rightPruned):
                    return True
            
            return False
        
        # edge cases
        if not root:
            return None
        
        prune(root)
        
        # is root guaranteed to be 1?
        # if not, we will have to update root as well after the traversal is complete.
        
        return root
    
        # Complexity
        # O(n) - visiting all nodes in the worst case
        # O(h) - height - cost of recursion
````
