### [Balanced Binary Tree](https://leetcode.com/problems/balanced-binary-tree/description/)

Given a binary tree, determine if it is height-balanced.

For this problem, a height-balanced binary tree is defined as:

> a binary tree in which the depth of the two subtrees of every node never differ by more than 1.

**Example 1:**

```
Given the following tree [3,9,20,null,null,15,7]:

    3
   / \
  9  20
    /  \
   15   7
Return true.
```

**Example 2:**

```
Given the following tree [1,2,2,3,3,null,null,4,4]:

       1
      / \
     2   2
    / \
   3   3
  / \
 4   4
Return false.
```

````python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    
    def isBalanced(self, root):
        """
        :type root: TreeNode
        :rtype: bool
        """
        
        # balanced = True if (height(left) - height(right)) <= 1 
        #            False otherwise
        # 
        # we can get height of every node. but don't have to recalculate
        # height eveyrtime. Calculating the height of the root will invoke
        # height calculation for every node. check the subtree heights for
        # every node and return false if it is not balanced.
        
        def findHeight(root):
            """ Returns the height and balance status for a given root """
            if not root:
                return 0, True
            
            leftHeight, leftBalanced = findHeight(root.left)
            rightHeight, rightBalanced = findHeight(root.right)
            
            rootHeight = 1 + max(leftHeight, rightHeight)
            
            if (not leftBalanced or not rightBalanced) or (abs(leftHeight - rightHeight) > 1):
                return rootHeight, False
            else:
                return rootHeight, True
        
        if not root:
            return True
        
        # using multiple return values here.
        # Alternate options are:
        #   pass a list with one element
        #   pass pointer/reference in C, C++ style
        _, rootBalanced = findHeight(root)
        
        return rootBalanced
        
    def isBalancedV1(self, root):
        """
        :type root: TreeNode
        :rtype: bool
        """
        
        # abs(height of leftSubtree - height of right subtree) <=1
        # find the height
        
        # cover the base cases
        
        if not root:
            return True
        
        # using state based solutions is not recommended at all.
        
        self.isTreeBalanced = True
        
        self.getHeight(root)
        
        return self.isTreeBalanced
 
    def getHeight(self, root):
        if not root:
            return 0
        
        # leaf node
        if not root.left and not root.right:
            return 1
    
        leftHeight = self.getHeight(root.left)
        rightHeight = self.getHeight(root.right)
        
        rootHeight = (1 + max(leftHeight, rightHeight))
        
        if abs(leftHeight - rightHeight) > 1:
            self.isTreeBalanced = False
        
        return rootHeight
````
