### [Merge Intervals](https://leetcode.com/problems/merge-intervals/)

iven an array of meeting time intervals consisting of start and end times [[s1,e1],[s2,e2],...] (si < ei), find the minimum number of conference rooms required.

**Example 1:**

```
Input: [[0, 30],[5, 10],[15, 20]]
Output: 2
```

**Example 2:**
```
Input: [[7,10],[2,4]]
Output: 1
```

````python
from typing import List

class Solution:
    def merge(self, intervals: List[List[int]]) -> List[List[int]]:
        # e.g. [1, 3], [2, 5] -> [1, 5]
        # [1, 5], [6, 8] -> [1, 5], [6, 8]
        # [1, 3], [2, 5], [4, 6], [8, 10] => [1, 6]
        #
        # Can I assume that the intervals are sorted?
        #
        #
        # condition for overlap
        #   t1, t2
        #   t2.start >= t1.start and t2.start <= t2.end
        #       // overlap here
        #   t3.start = min(t1.start, t2.start)
        #   t3.end = max(t1.end, t2.end)
        #
        # move t3 to merged_intervals
        #
        # for the next round,
        # t1 = t3
        # t2 = next_t in line
        #
        #
        # sort the intervals
        # push the first into merged list
        #
        # for intervals from 1..end
        #   t1 = merged[-1]
        #   t2 = cur_interval
        #   if t1 & t2 overlaps:
        #       update t1 with new values
        #   else:
        #       add t2 to merged list
        
        # edge cases
        # 
        if not intervals:
            return []
        
        # assume that input lists are valid. so no need of error checking
        
        intervals.sort(key=lambda x: x[0])
        
        merged_intervals = [intervals[0]]
        
        for i in range(1, len(intervals)):
            t1 = merged_intervals[-1]
            t2 = intervals[i]
            if t2[0] >= t1[0] and t2[0] <= t1[1]:
                # we have an overlap
                t1[1] = max(t1[1], t2[1])
                # t1[0] must be minimum here because we already sorted. 
            else:
                merged_intervals.append(t2)
        
        return merged_intervals
````
