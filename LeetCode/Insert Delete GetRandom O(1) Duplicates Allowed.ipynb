### [Insert Delete GetRandom O(1) Duplicates Allowed](https://leetcode.com/problems/insert-delete-getrandom-o1-duplicates-allowed/)

Design a data structure that supports all following operations in average O(1) time.

**Note: Duplicate elements are allowed.**

- insert(val): Inserts an item val to the collection.
- remove(val): Removes an item val from the collection if present.
- getRandom: Returns a random element from current collection of elements. The probability of each element being returned is linearly related to the number of same value the collection contains.

Example:
```
// Init an empty collection.
RandomizedCollection collection = new RandomizedCollection();

// Inserts 1 to the collection. Returns true as the collection did not contain 1.
collection.insert(1);

// Inserts another 1 to the collection. Returns false as the collection contained 1. Collection now contains [1,1].
collection.insert(1);

// Inserts 2 to the collection, returns true. Collection now contains [1,1,2].
collection.insert(2);

// getRandom should return 1 with the probability 2/3, and returns 2 with the probability 1/3.
collection.getRandom();

// Removes 1 from the collection, returns true. Collection now contains [1,2].
collection.remove(1);

// getRandom should return 1 and 2 both equally likely.
collection.getRandom();
```

````python
import random

class RandomizedCollection(object):

    def __init__(self):
        """
        Initialize your data structure here.
        """
        # O(1) time for
        #   insert
        #   remove
        #   getRandom - each element returned should have the same probability
        
        
        # This is the continuation of https://leetcode.com/problems/insert-delete-getrandom-o1/
        
        self.container = []
        self.containerMap = {}
        

    def insert(self, val):
        """
        Inserts a value to the set. Returns true if the set did not already contain the specified element.
        :type val: int
        :rtype: bool
        """
        # to support duplicate entries, we could store the set of indexes
        # of the val in our map. why set instead of list? During deletion,
        # we need to selectively remove the last index the indexes of replacement.
        # we have to use linear search if we use list, and that will not give
        # us O(1) removal time.
        # upon insertion we add the new index to our map
        # upon deletion, we remove the val fro the map only if all the indexes are removed
        #   it doesn't matter which duplicate entry we remove. so any index of that val
        #   can be removed.
        # upon getrandom(), we will get equal probability (ie.num-entries-of-val/total-entries)
        # only if we keep duplicates in our container
        
        
        # add val to our container
        valIndex = len(self.container)
        self.container.append(val)
        
        # add val and valIndex to our map
        if val not in self.containerMap:
            # create the set if this is the first entry for the val
            self.containerMap[val] = set()
        
        self.containerMap[val].add(valIndex)
        
         # new entry will have only 1 index
        return (len(self.containerMap[val]) == 1)
            
        
    def remove(self, val):
        """
        Removes a value from the set. Returns true if the set contained the specified element.
        :type val: int
        :rtype: bool
        """
    
        if val not in self.containerMap:
            return False
        
        # make sure list is not empty
        # if val is in containerMap then list should also be not empty
        valIndexes = self.containerMap[val]
        
        # pop a random index
        delIndex = valIndexes.pop()
        
        # move the last element to the deleted position
        replacement, replacementIndex = self.container[-1], len(self.container) - 1
        self.container[delIndex] = replacement
        
        
        # if the containerMap has a valid entry for lastIndex
        # of the last item, then remove it.
        if self.containerMap[replacement]:
            # now add the deleted index to the index list of last element
            self.containerMap[replacement].add(delIndex)

            # now we should find the slot of last val in our map
            # and remove the lastIndex from the index list
            self.containerMap[replacement].discard(replacementIndex)
        
        # drop the last element from our container.
        self.container.pop()
        
        # remove the deleted val from our map as well
        # after all the instances of 'val' are removed.
        if not valIndexes:
            del self.containerMap[val]
        
        return True

    def getRandom(self):
        """
        Get a random element from the set.
        :rtype: int
        """
        if self.container:
            # This still holds good, even with duplicates
            return random.choice(self.container)
        else:
            raise IndexError


# Your RandomizedCollection object will be instantiated and called as such:
# obj = RandomizedCollection()
# param_1 = obj.insert(val)
# param_2 = obj.remove(val)
# param_3 = obj.getRandom()
````
