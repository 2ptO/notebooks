## [Four Sum] (https://leetcode.com/problems/4sum/description/_

```python
class SolutionBruteForce(object):
    def fourSum(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: List[List[int]]
        """
        
        # solution set must contain unique quadruplets.
        # can in the input nums contain duplicate numbers?
        # 
        # a + b + c + d = target
        # can be split into 2 two-sum problem?
        # target - (a + b), are there numbers in nums that matches this value?
        #
        # pick 1, 0 {how many numbers is {-1, 0, -2, 2} that form -1?}
        #
        # pick a combination of 2 from nums
        #   find 2 num from nums without the chosen two
        # 
        # two sum problem.. can find whther two numbers that add up to a target
        # exist or not. using a set.. can do that in O(n) time. collect the set
        # of numbers that sum up to the given target.
        #
        # 
        def twoSum(nums, target, ignoreIndices):
            seen = set()
            pairs = []
            for index, num in enumerate(nums):
                if index in ignoreIndices:
                    continue

                if target-num in seen:
                    pairs.append([num, (target-num)])
                else:
                    seen.add(num)

            return pairs
        
        solutionSet = {}
        
        # edge cases
        # empty nums?
        #   taken care
        
        for i in range(len(nums)):
            for j in range(i+1, len(nums)):
                for pair in twoSum(nums, target-(nums[i] + nums[j]), {i, j}):
                    # print(pair, "for {} and {}".format(nums[i], nums[j]))
                    solution = sorted([nums[i], nums[j]] + pair)
                    hashSolution = ""
                    for num in solution:
                        hashSolution += (str(num) + "*")

                    if hashSolution not in solutionSet:
                        solutionSet[hashSolution] = solution
                    
        return solutionSet.values()
```

```python
class Solution(object):
    def fourSum(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: List[List[int]]
        """
        
        def twoSumPairs(nums, target):
            result = []
            
            # using two pointer method since the input
            # nums are given in sorted order.
            
            start, end = 0, len(nums) - 1
            
            while start < end:
                curSum = nums[start] + nums[end]
                if curSum < target:
                    # we need add more to the curSum to meet the target
                    # since the input nums are sorted, we can move further right
                    # to add nums with higher values
                    start += 1
                elif curSum > target:
                    # we need to reduce curSum to meet the target
                    # since the input nums are sorted, we can move further left
                    # to add nums with lower values
                    end -= 1
                else:
                    # met the target
                    result.append([nums[start], nums[end]])
                    # move both start and end as we found the target with
                    # current start and end
                    start += 1
                    end -= 1
                    
                    # in order to avoid duplicate values in the results, we can
                    # skip over numbers which are repeated in the input nums
                    # since our input nums is sorted, repeated numbers will be
                    # placed adjacently.
                    
                    # skip duplicates from the current start
                    while start < end and nums[start] == nums[start - 1]:
                        start += 1
                    
                    # skip duplicates from the current end
                    while end > start and nums[end] == nums[end + 1]:
                        end -= 1
            
            return result
        
        
        def threeSum(nums, target, numsAreSorted = True):
            """
            :type nums: List[int]
            :rtype: List[List[int]]
            """

            # we want a minimum of three numbers (i.e. cur_number + two more numbers
            # from the input nums). so iterate on the nums until the last two numbers

            # threeSum problem can be reduced to num1 + twoSum problem, where we find
            # twosum excuding num1. target for twoSum woudl then be target-num1

            if not nums or len(nums) < 3:
                return []

            # sort the nums first 
            if not numsAreSorted:
                nums.sort()

            result = []
            for i in range(len(nums) - 2):
                # we need this check to avoid duplicate results
                if i > 0 and nums[i] == nums[i-1]:
                    # nums[i] same as previous number
                    # we already solved that. so continue looking
                    # for the next number
                    continue

                twoSumTarget = target - nums[i]
                for pair in twoSumPairs(nums[i+1:], twoSumTarget):
                    result.append([nums[i]] + pair)

            return result
        
        def findKSum(nums, K, target):
            result = []
            
            if not nums or len(nums) < K:
                # invalid input length
                return result
            
            lowestPossibleTarget = nums[0] * K #
            highestPossibleTarget = nums[-1] * K
            
            if target < lowestPossibleTarget or target > highestPossibleTarget:
                # target out of bounds for the current sorted numbers
                return result
            
            # assume that the numbers are already sorted.
            # nums.sort()
            
            if K < 3:
                return twoSumPairs(nums, target)
            
            for i in range(len(nums) - (K-1)):
                if i > 0 and nums[i] == nums[i-1]:
                    continue
                
                targetToKMinusOne = target - nums[i]
                for numForKMinusOneSum in findKSum(nums[i+1:], K-1, targetToKMinusOne):
                    result.append([nums[i]] + numForKMinusOneSum)
            
            return result
        
        if not nums:
            return []
        
        # sort only once
        nums.sort()
        return findKSum(nums, 4, target)

        # Extending from my solutin to threesum
        # we want a minimum of 4 numbers here 
        # take every number from input nums and find three other nums using threeSum
        # since we need a minimum of three nums for threeSum(), we can stop
        # iterating the input list when we reach less than three in the remaining list of numbers
        
#         result = []
        
#         if not nums or len(nums) < 4:
#             return []
        
#         nums.sort()
        
#         for i in range(len(nums) - 3):
#             # like three sum, avoid duplicates here as well
#             if i > 0 and nums[i] == nums[i-1]:
#                 continue
            
#             threeSumTarget = target - nums[i]
#             for threeSumNum in threeSum(nums[i+1:], threeSumTarget):
#                 result.append([nums[i]] + threeSumNum)
            
#         return result
```
