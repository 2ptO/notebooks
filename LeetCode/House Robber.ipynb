### House Robber

From https://leetcode.com/problems/house-robber/description/

ou are a professional robber planning to rob houses along a street. Each house has a certain amount of money stashed, the only constraint stopping you from robbing each of them is that adjacent houses have security system connected and it will automatically contact the police if two adjacent houses were broken into on the same night.

Given a list of non-negative integers representing the amount of money of each house, determine the maximum amount of money you can rob tonight without alerting the police.

Example 1:
```
Input: [1,2,3,1]
Output: 4
Explanation: Rob house 1 (money = 1) and then rob house 3 (money = 3).
             Total amount you can rob = 1 + 3 = 4.
```

Example 2:
```
Input: [2,7,9,3,1]
Output: 12
Explanation: Rob house 1 (money = 2), rob house 3 (money = 9) and rob house 5 (money = 1).
             Total amount you can rob = 2 + 9 + 1 = 12.
```


````python
class Solution(object):
    def rob(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        # Its a maximization problem
        # cannot rob adjacent houses
      
        # single house
        
        # two house
        
        # > 2 houses
        # rob the third only if 1st & 3rd > max(1st, 2nd)
        # similarly, rob the fourth if 2nd & 4th > max(1st, 2nd, 3rd)
        # max_rob_so_far is needed
        
        if not nums:
            return 0
        
        if len(nums) == 1:
            return nums[0]
        
        if len(nums) == 2:
            return max(nums[0], nums[1])
        
         
        # writing the problem statement in a formula like stmt really helps.
        # f(i) = max(f(i-2) + Num_i, f(i-1))
        # f(i-1) - max profit until the previous house
        # f(i-2) - max profit until the house before the previous house
        
        max_rob_from_prev_house = 0 # house before the first house
        max_rob_from_prev_prev_house = 0 # house before the previous house of the first house
        
        for n in nums:
            max_rob_from_this_house = max(max_rob_from_prev_prev_house + n, max_rob_from_prev_house)
            max_rob_from_prev_prev_house = max_rob_from_prev_house
            max_rob_from_prev_house = max_rob_from_this_house
        
        # exhausted all houses. Return the max robbed from the last house.
        return max_rob_from_prev_house
        
    
    def robWithAddlSpace(self, nums):
        # Runs in O(n) time but takes O(n) space
    
        if not nums:
            return 0
        
        if len(nums) == 1:
            return nums[0]
        
        if len(nums) == 2:
            return max(nums[0], nums[1])
         
        max_rob_at_house = [0 for _ in range(len(nums))]
        max_rob_at_house[0] = nums[0]
        max_rob_at_house[1] = max(nums[0], nums[1])
        
        
        for i in range(2, len(nums)):
            max_rob_at_house[i] = max(max_rob_at_house[i-2] + nums[i], max_rob_at_house[i-1])
        
        # return the max profit from the last house8
        return max_rob_at_house[-1]
````

````python
s = Solution()
assert (s.rob([]) == 0)
assert (s.rob([10]) == 10)
assert (s.rob([4, 5]) == 5)
assert (s.rob([1,2,3,1]) == 4)
assert (s.rob([2,7,9,3,1]) == 12)
assert (s.rob([2,0, 3, 3, 0, 2, 10, 31, 4, 11]) == 49)

assert (s.robWithAddlSpace([]) == 0)
assert (s.robWithAddlSpace([10]) == 10)
assert (s.robWithAddlSpace([4, 5]) == 5)
assert (s.robWithAddlSpace([1,2,3,1]) == 4)
assert (s.robWithAddlSpace([2,7,9,3,1]) == 12)
assert (s.robWithAddlSpace([2,0, 3, 3, 0, 2, 10, 31, 4, 11]) == 49)
````
