### [Sqrt(x)](https://leetcode.com/problems/sqrtx/)

Implement int sqrt(int x).

Compute and return the square root of x, where x is guaranteed to be a non-negative integer.

Since the return type is an integer, the decimal digits are truncated and only the integer part of the result is returned.

**Example 1:**
```
Input: 4
Output: 2
```

**Example 2:**
```
Input: 8
Output: 2
```
Explanation: The square root of 8 is 2.82842..., and since 
             the decimal part is truncated, 2 is returned.

````python
class Solution(object):
    def mySqrt(self, x):
        """
        :type x: int
        :rtype: int
        """
        # Compute square root of x
        # [1, 2, 3, 4, 5, 6, 7...]
        # [1, 4, 9, 16, 25, 36, 49...x]
        
        # we want to value from list 1 based on binary search in list2
        # if 'x' is not perfect square, then we pick the value which
        # exists to left of 'x'

        if x in [0, 1]:
            return x
        
        low, high = 1, x
        
        while low <= high:
            mid = low + (high-low) // 2
            
            # just like regular binary search, compare the key (x) with
            # the mid point.. with a twist. our mid point is square of mid point
            # as thats the value we are looking to find.
            
            if (x == mid * mid):
                return mid
            elif (x < mid * mid):
                high = mid - 1
            else:
                low = mid + 1
        
        # we crossed the low & high barrier.
        # happens only if we don't find the exact sqrt.
        # so return 1 below the next value
        return low-1
    
    def mySqrtMath(self, x):
        """
        :type x: int
        :rtype: int
        """
        # option 2:
        # high school math
        #   reduce x until the square root is found
        root = x
        while root * root > x:
            root = (root + x/root) // 2
        
        return root
    
        
````

````python
tests = {
    "test" : [
        {
            "x": 36,
            "sqrt": 6
        },
        {
            "x": 35,
            "sqrt": 5
        },
        {
            "x": 1024,
            "sqrt": 32
        },
        {
            "x": 5,
            "sqrt": 2
        },
        {
            "x": 37,
            "sqrt": 6
        },
        {
            "x": 4352,
            "sqrt": 65
        },
        
    ]
}
````

````python
s = Solution()

for test in tests["test"]:
    assert(s.mySqrt(test["x"]) == test["sqrt"])
````
