### [Reverse words in a string ii](https://leetcode.com/problems/reverse-words-in-a-string-ii/)

Given an input string , reverse the string word by word. 

**Example:**

```
Input:  ["t","h","e"," ","s","k","y"," ","i","s"," ","b","l","u","e"]
Output: ["b","l","u","e"," ","i","s"," ","s","k","y"," ","t","h","e"]
```

**Note:**

- A word is defined as a sequence of non-space characters.
- The input string does not contain leading or trailing spaces.
- The words are always separated by a single space.

**Follow up:** Could you do it in-place without allocating extra space?



````python
class Solution(object):
    def reverseWords(self, strings):
        """
        :type str: List[strings]
        :rtype: void Do not return anything, modify str in-place instead.
        """
        # understanding the problem:
        # not full string reversal
        # reverse the string in a way that only the words within the string are reversed
        # no leading and trailing spaces
        # words are always separated by single space
        # e.g.
        #   the sky is blue
        #   blue is sky the
        # 
        # intuition:
        # condition: reverse in place
        # brute force first.
        #   starting with new string
        #   separate by words
        #   new_string = word + (if newstring " " else "") + new_string
        #
        # to do in place:
        # reverse the string
        #   reverse each word?
        # the sky is blue
        # eulb si yks eht
    
        
        def reverse(string_list, start, end):
            # python lists are passed by reference.
            # changing the order of string_list will affect the source reference too.
            while start < end:
                string_list[start], string_list[end] = string_list[end], string_list[start]
                start += 1
                end -= 1
        
        # edge cases
        # empty list
        # list with only one string
        # no leading or trainling spaces..so no cases like ["a", " "]
        if not strings or len(strings) < 2:
            return
        
        # reverse the string first
        reverse(strings, 0, len(strings) - 1)
        
        # now reverse individual words
        start = 0
        delim = 1
        
        while delim <= len(strings):
            # account for last word as well.
            # we reverse the word under either of two conditions:
            #   reached the end..so do not check for space
            #   current char is a space
            if delim == len(strings) or strings[delim].isspace():
                # reverse the word from start to delim
                reverse(strings, start, delim-1)
                start = delim + 1
            
            delim += 1
            
````
