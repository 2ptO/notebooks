### [Sum Root to Leaf Numbers](https://leetcode.com/problems/sum-root-to-leaf-numbers/)


Given a binary tree containing digits from 0-9 only, each root-to-leaf path could represent a number.

An example is the root-to-leaf path 1->2->3 which represents the number 123.

Find the total sum of all root-to-leaf numbers.

Note: A leaf is a node with no children.

**Example:**
```
Input: [1,2,3]
    1
   / \
  2   3
Output: 25
Explanation:
The root-to-leaf path 1->2 represents the number 12.
The root-to-leaf path 1->3 represents the number 13.
Therefore, sum = 12 + 13 = 25.
```

**Example 2:**
```
Input: [4,9,0,5,1]
    4
   / \
  9   0
 / \
5   1
Output: 1026
```
Explanation:
The root-to-leaf path 4->9->5 represents the number 495.
The root-to-leaf path 4->9->1 represents the number 491.
The root-to-leaf path 4->0 represents the number 40.
Therefore, sum = 495 + 491 + 40 = 1026.


````python
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None

class Solution:
    def sumNumbers(self, root):
        """
        :type root: TreeNode
        :rtype: int
        """
        # Generate all paths from the root
        # sum values of each path

#         at each node
#         when traversing the child node, pass the path traversed so far
#         sum only when reaching a leaf node or add path to a list.. and then
#         sum them up totally.
#         sum = 0
        
#         node = 9, path = 4
#         node = 5, path = 49
#             leaf node, so sum = 4
#         node = 1, path = 49
        
        all_paths = []
        self.find_paths(root, 0, all_paths)
        return sum(all_paths)
    
    def find_paths(self, node, path, all_paths):
        # if leaf node, add path to all_paths
        if node:
            path = path*10 + node.val
            if node.left == None and node.right == None:
                all_paths.append(path)
            else:
                self.find_paths(node.left, path, all_paths)
                self.find_paths(node.right, path, all_paths)
````
