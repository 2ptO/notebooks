### [Minimum Depth of a Binary Tree](https://leetcode.com/problems/minimum-depth-of-binary-tree/description/)

Given a binary tree, find its minimum depth.

The minimum depth is the number of nodes along the shortest path from the root node down to the nearest leaf node.

Note: A leaf is a node with no children.

**Example**:
```
Given binary tree [3,9,20,null,null,15,7],

    3
   / \
  9  20
    /  \
   15   7
return its minimum depth = 2.
```

````python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

from collections import deque

class Solution:
    def minDepth(self, root):
        """
        :type root: TreeNode
        :rtype: int
        """
        
        return self.minDepthDFS(root)
    
    def minDepthDFS(self, root):
        """
        :type root: TreeNode
        :rtype: int
        """
        # with DFS, it is not guaranteed to hit the closest leaf node first.
        # so we will have to keep track of minimum depth.
        # with tree, most problems can be solved as subproblems.
        # for a tree, its min depth will be
        #   1 (for itself) + minDepth(of its left and right subtrees)
        
        if not root:
            return 0
        
        if not root.left:
            # only right exists
            return 1 + self.minDepthDFS(root.right)
        elif not root.right:
            # only left subtree exists. so get its min depth
            return 1 + self.minDepthDFS(root.left)
        else:
            # both left and right exists. check for both.
            return 1 + min(self.minDepthDFS(root.left), self.minDepthDFS(root.right))
    
    def minDepthBFS(self, root):
        """
        :type root: TreeNode
        :rtype: int
        """
        # shortest path.. so use BFS
        # do a bread first search. return the depth upon hitting the first leaf node.
        
        # edge cases first
        if not root:
            return 0 # not sure if LC considers this as or or 1. IMO, it should be considered as 0
        
        # bfs needs a queue.. root is at depth 1.
        depth = 1
        queue = deque([(root, depth)])
        
        while queue:
            node, depth = queue.popleft()
            
            if not node.left and not node.right:
                # found a leaf node.
                return depth
            
            if node.left:
                queue.append((node.left, depth + 1))
            if node.right:
                queue.append((node.right, depth + 1))
        
        # if last node is the closest leaf node, return the final depth here.
        # Thinking again, we shouldn't really come here at all
        print("Returning depth ", depth, "after draining the queue")
        return depth
    
````
