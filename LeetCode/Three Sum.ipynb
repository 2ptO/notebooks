## [Three Sum](https://leetcode.com/problems/3sum/description/)

Given an array nums of n integers, are there elements a, b, c in nums such that a + b + c = 0? Find all unique triplets in the array which gives the sum of zero.

Note:

The solution set must not contain duplicate triplets.

Example:
```
Given array nums = [-1, 0, 1, 2, -1, -4],

A solution set is:
[
  [-1, 0, 1],
  [-1, -1, 2]
]
```

````python
class Solution(object):
    def threeSum(self, nums):
        """
        :type nums: List[int]
        :rtype: List[List[int]]
        """
        
        def twoSumPairs(nums, target):
            result = []
            
            # using two pointer method since the input
            # nums are given in pre-sorted by the caller
            
            start, end = 0, len(nums) - 1
            
            while start < end:
                curSum = nums[start] + nums[end]
                if curSum < target:
                    # we need add more to the curSum to meet the target
                    # since the input nums are sorted, we can move further right
                    # to add nums with higher values
                    start += 1
                elif curSum > target:
                    # we need to reduce curSum to meet the target
                    # since the input nums are sorted, we can move further left
                    # to add nums with lower values
                    end -= 1
                else:
                    # met the target
                    result.append([nums[start], nums[end]])
                    # move both start and end as we found the target with
                    # current start and end
                    start += 1
                    end -= 1
                    
                    # in order to avoid duplicate values in the results, we can
                    # skip over numbers which are repeated in the input nums
                    # since our input nums is sorted, repeated numbers will be
                    # placed adjacently.
                    
                    # skip duplicates from the current start
                    while start < end and nums[start] == nums[start - 1]:
                        start += 1
                    
                    # skip duplicates from the current end
                    while end > start and nums[end] == nums[end + 1]:
                        end -= 1
            
            return result
        
        # we want a minimum of three numbers (i.e. cur_number + two more numbers
        # from the input nums). so iterate on the nums until the last two numbers

        # threeSum problem can be reduced to num1 + twoSum problem, where we find
        # twosum excuding num1. target for twoSum woudl then be target-num1
        
        if not nums or len(nums) < 3:
            return []
        
        # sort the nums first 
        nums.sort()
        
        result = []
        target = 0
        
        for i in range(len(nums) - 2):
            # we need this check to avoid duplicate results
            if i > 0 and nums[i] == nums[i-1]:
                # nums[i] same as previous number
                # we already solved that. so continue looking
                # for the next number
                continue
                
            twoSumTarget = target - nums[i]
            for pair in twoSumPairs(nums[i+1:], twoSumTarget):
                result.append([nums[i]] + pair)

        return result
````
