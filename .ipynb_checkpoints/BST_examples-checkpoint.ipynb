In order successor of node in a binary tree


```python
def find_parent(root, node):
    if root and root != node:
        # check this node
        if root.left == node or root.right == node:
            return root
        
        # find parent is in left subtree or not
        parent = find_parent(root.left, node)
        if parent:
            return parent
        else:
            return find_parent(root.right, node)
    return None

def inorder_successor(root, p):
    if p.right:
        # return left most node in the right subtree
        io_successor = p.right
        while io_successor.left:
            io_successor = io_successor.left

        return io_successor

    parent = find_parent(root, p)
    if parent:
        if parent.left == p:
            return parent
        else:
            parent.right = None
            io_successor = inorder_successor(root, parent)
            parent.right = p
            return io_successor
    else:
        return None
```
