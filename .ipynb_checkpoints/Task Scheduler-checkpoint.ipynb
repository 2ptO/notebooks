```python
# Question from https://leetcode.com/problems/task-scheduler/description/

from collections import Counter
class Solution(object):
    def leastInterval(self, tasks, n):
        """
        :type tasks: List[str]
        :type n: int
        :rtype: int
        """
        # 6:15
        # [A, A, A, B, B, B]
        # tasks can be in any order
        # must complete all tasks
        # 
        # tasks A .. Z
        # Schedule a task
        #   if last task != current task and cooling_timer_not_expired for current task
        # map[A..Z]
        # cooling[A] = interval
        # A - prev_task = A, curr_task = A, interval = 1
        # next in line is A.. can't pickup A.. 
        # next in line is A..can't pickup.. A
        # next in line is B, 
        #   B, last task != current task, B is cooling? no
        #   pick up B.. internvals = 2
        # next is B... can't pick up
        # next is also B.. can't pick up
        # reached the end. 
        # wait.. idle += 1, interval = 3
        # pick up A..
        #
        # when to stop looking for jobs?
        # which data structure to use to service the tasks?
        # queue.. but question says task can be any order.  priority queue?
        
        #
        # if using priority queue, what is considered as the priority?
        # cooling interval?
        # 
        
        # edge cases
        if not tasks:
            return 0
        
        tasksMap = Counter(tasks) # map of tasks to count and interval
        coolingMap = {}
        tasksRemaining = True
        
        intervals = 0
        while tasksRemaining:
            # is there a next job?
            tasksRemaining = False
            nextTask = None
            for task, taskCnt in tasksMap.most_common():
                if taskCnt > 0:
                    tasksRemaining = True
                    taskIsCooling = task in coolingMap and (intervals - coolingMap[task] < n)
                    if taskIsCooling:
                        # look for next available task
                        continue
                    nextTask = task
                    break
            
            if nextTask:
                intervals += 1
                coolingMap[nextTask] = intervals
                tasksMap[nextTask] -= 1
            elif tasksRemaining:
                # some tasks remaining, but nothing ready to be scheduled.
                # stay idle
                intervals += 1
        
        return intervals
```
