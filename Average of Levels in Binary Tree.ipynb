### [Average of levels in Binary Tree](https://leetcode.com/problems/average-of-levels-in-binary-tree/description/)

Given a non-empty binary tree, return the average value of the nodes on each level in the form of an array.
Example 1:
```
Input:
    3
   / \
  9  20
    /  \
   15   7
Output: [3, 14.5, 11]
Explanation:
The average value of nodes on level 0 is 3,  on level 1 is 14.5, and on level 2 is 11. 
Hence return [3, 14.5, 11].
```

````python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None
from collections import deque

class Solution:
    
    def averageOfLevels(self, root):
        """
        :type root: TreeNode
        :rtype: List[float]
        """
        return self.averageOfLevelsDFS(root)


    # how did I miss this logic? I have used this several times in the past
    # this one ran faster than all other solution
    def averageOfLevelsBFSv2(self, root):
        """
        :type root: TreeNode
        :rtype: List[float]
        """
        
        if not root:
            return []
        
        queue = deque()
        queue.append(root)
        
        averages = []
        
        while queue:
            sumAtLevel = 0
            count = len(queue)
            i = 0
            while i < count:
                node = queue.popleft()
                sumAtLevel += node.val
                
                if node.left:
                    queue.append(node.left)
                if node.right:
                    queue.append(node.right)
                
                i += 1
            
            averages.append(sumAtLevel/count)
        
        return averages
    
    
    def averageOfLevelsDFS(self, root):
        """
        :type root: TreeNode
        :rtype: List[float]
        """
        
        def dfs(node, level, levelSums, numNodes):
            if node:
                # process the node
                if level < len(levelSums):
                    levelSums[level] += node.val
                    numNodes[level] += 1
                else:
                    levelSums.append(node.val)
                    numNodes.append(1)
                    
                dfs(node.left, level + 1, levelSums, numNodes)
                dfs(node.right, level + 1, levelSums, numNodes)
        
        levelSums = []
        numNodes = []
        
        dfs(root, 0, levelSums, numNodes)
        
        return [levelSums[i]/numNodes[i] for i in range(len(levelSums))]
        
        
    def averageOfLevelsBFS(self, root):
        """
        :type root: TreeNode
        :rtype: List[float]
        """
        
        # LEVEL order traversal
        # sum up all nodes at a given level.
        # find the average before jumping to new level.
        # since we use level order traversal, we always go 
        # sequentially.
        
        if not root:
            return []
        
        queue = deque()
        queue.append((root, 0))
        
        # Return value
        levelAverages = []
        sumAtLevel, numNodes = 0, 0
        
        while queue:
            node, level = queue.popleft()
            
            # Process the node
            if level > len(levelAverages):
                levelAverages.append(sumAtLevel / numNodes)
                sumAtLevel, numNodes = 0, 0
            
            sumAtLevel += node.val
            numNodes += 1
            
            if node.left:
                queue.append((node.left, level + 1))
            if node.right:
                queue.append((node.right, level + 1))
        
        if numNodes > 0:
            # count the nodes at the last level
            levelAverages.append(sumAtLevel/numNodes)
            
        return levelAverages
````
